### We can see price is dependent on other factors, but id, date, zipcode etc. does'nt impact price. So dropping not needed columns

```python
df1=df.drop(['id','date','yr_built','yr_renovated','zipcode'],axis=1)
```

```python
df1.head()
```

```python
df1['grade'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df1.drop(['price'],axis=1)
y=df1['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.9,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
ran=RandomForestRegressor(n_estimators=400)
```

```python
ran.fit(x_train,y_train)
```

```python
ran.score(x_test,y_test)
```

```python
ran.predict([x.iloc[0]])
```
